## Задача 1

Да се намери $n$ така, че $n$-тата съставна
квадратурна формула на **правоъгълниците**/**трапеците**/**Симпсън**
да приближава интеграла
$\displaystyle
I = \int_{0}^{1} \frac{dx}{1+x}
$ с грешка, не по-голяма от $\displaystyle \varepsilon = 0.00001.$

Да се намери това приближение.


```python
import numpy as np
import sympy as sp

# rectangles, first we calc for n, then =>
n_rect = 92
def f(x):
    return 1 / (1 + x)

a = 0
b = 1
xmesh_rect = np.linspace(a, b, n_rect + 1)

approximate_integral_rect = 0
for i in range(1, n_rect + 1):
    approximate_integral_rect += f((xmesh_rect[i - 1] + xmesh_rect[i]) / 2)
    
approximate_integral_rect *= (b - a) / n_rect
print(approximate_integral_rect)

x = sp.symbols('x')
# print(sp.integrate(1/(1+x), (x,0,1))) # log2
print(np.log(2))

# simpson n = (10**5/120)**(1/4)
n_simp = 6
xmesh_simp = np.linspace(a, b, n_simp + 1)
approximate_integral_simp = 0
for i in range(1, n_simp + 1):
    approximate_integral_simp += f(xmesh_simp[i-1]) + 4*f((xmesh_simp[i-1] + xmesh_simp[i])/2) + f(xmesh_simp[i])
    
approximate_integral_simp *= (b - a) / (6 * n_simp)
print(approximate_integral_simp)
```

### Задача 2

По метода на неопределените коефициенти да се намери формулата на Гаус-Льожандър с два възела.

```python
# from seminar
x = sp.symbols('x')
x1 = sp.symbols('x1')
x2 = sp.symbols('x2')
A1 = sp.symbols('A1')
A2 = sp.symbols('A2')
sp.solve(
    [
        sp.Eq(sp.integrate(1, (x,-1,1)), A1 + A2),
        sp.Eq(sp.integrate(x, (x,-1,1)), A1*x1 + A2*x2),
        sp.Eq(sp.integrate(x**2, (x,-1,1)), A1*x1**2 + A2*x2**2),
        sp.Eq(sp.integrate(x**3, (x,-1,1)), A1*x1**3 + A2*x2**3)
    ]
)
```

### Формули на Гаус-Льожандър:

| Брой възли | Коефициенти                 | Възли                          | АСТ |
|-----------|-----------------------------|--------------------------------|-----|
| 1         | A₁ = 2                      | x₁ = 0                         | 1   |
| 2         | A₁ = 1<br>A₂ = 1            | x₁ = −1/√3<br>x₂ = 1/√3        | 3   |
| 3         | A₁ = 5/9<br>A₂ = 8/9<br>A₃ = 5/9 | x₁ = −√(3/5)<br>x₂ = 0<br>x₃ = √(3/5) | 5   |
