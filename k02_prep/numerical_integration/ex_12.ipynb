## Задача 1

Да се намери $n$ така, че $n$-тата съставна
квадратурна формула на **правоъгълниците**/**трапеците**/**Симпсън**
да приближава интеграла
$\displaystyle
I = \int_{0}^{1} \frac{dx}{1+x}
$ с грешка, не по-голяма от $\displaystyle \varepsilon = 0.00001.$

Да се намери това приближение.


```python
import numpy as np
import sympy as sp

# rectangles, first we calc for n, then =>
n_rect = 92
def f(x):
    return 1 / (1 + x)

a = 0
b = 1
xmesh_rect = np.linspace(a, b, n_rect + 1)

approximate_integral_rect = 0
for i in range(1, n_rect + 1):
    approximate_integral_rect += f((xmesh_rect[i - 1] + xmesh_rect[i]) / 2)
    
approximate_integral_rect *= (b - a) / n_rect
print(approximate_integral_rect)

x = sp.symbols('x')
# print(sp.integrate(1/(1+x), (x,0,1))) # log2
print(np.log(2))

# simpson n = (10**5/120)**(1/4)
n_simp = 6
xmesh_simp = np.linspace(a, b, n_simp + 1)
approximate_integral_simp = 0
for i in range(1, n_simp + 1):
    approximate_integral_simp += f(xmesh_simp[i-1]) + 4*f((xmesh_simp[i-1] + xmesh_simp[i])/2) + f(xmesh_simp[i])
    
approximate_integral_simp *= (b - a) / (6 * n_simp)
print(approximate_integral_simp)
```

### Допълнителна задача

Да се намери $n$ така, че $n$-тата съставна
квадратурна формула на **Симпсън** да приближава интеграла 
$\displaystyle I=\int_{0}^{1} \sin x ~dx$ с грешка, не по-голяма от $\epsilon =0.001$. Да се намери приближението.

```python
# симпсън
import sympy as sp
import numpy as np

x = sp.Symbol('x')
b = 1
a = 0

def f(x):
    return sp.sin(x)

def f_num(x):
    return np.sin(x)

# намираме четвъртата производна
dddd_f = sp.diff(sp.diff(sp.diff(sp.diff(f(x), x), x), x), x)
print(dddd_f)

# трябва да сметнем n
# може на хартия може и тук
space = np.linspace(0, 1, 1000)
dddd_f_num = sp.lambdify(x, dddd_f, 'numpy')
max_f4 = np.max(np.abs(dddd_f_num(space)))

epsilon = 0.001

# без минус, взима се модулът
n = int(np.ceil(np.power((b-a)**5 * max_f4 / (2880 * epsilon), 1/4)))
print(int(n))
print(f"error: {(b-a)**5 * max_f4 / (2880 * n ** 4):.10f}")

# стандартно смятаме интегралното приближение
x_mesh = np.linspace(a, b, n + 1)
app_val = 0

for i in range(1, n + 1):
    app_val += f_num(x_mesh[i-1]) + 4 * f_num((x_mesh[i-1] + x_mesh[i])/2)+ f_num(x_mesh[i])
    
app_val *= (b-a)/(6*n)
print("approximate: ", app_val) 
print("real: ",sp.integrate(f(x), (x, a, b)).evalf()) 
```

### Задача 2

По метода на неопределените коефициенти да се намери формулата на Гаус-Льожандър с два възела.

```python
# from seminar
x = sp.symbols('x')
x1 = sp.symbols('x1')
x2 = sp.symbols('x2')
A1 = sp.symbols('A1')
A2 = sp.symbols('A2')
sp.solve(
    [
        sp.Eq(sp.integrate(1, (x,-1,1)), A1 + A2),
        sp.Eq(sp.integrate(x, (x,-1,1)), A1*x1 + A2*x2),
        sp.Eq(sp.integrate(x**2, (x,-1,1)), A1*x1**2 + A2*x2**2),
        sp.Eq(sp.integrate(x**3, (x,-1,1)), A1*x1**3 + A2*x2**3)
    ]
)
```

### Формули на Гаус-Льожандър:

| Брой възли | Коефициенти                 | Възли                          | АСТ |
|-----------|-----------------------------|--------------------------------|-----|
| 1         | A₁ = 2                      | x₁ = 0                         | 1   |
| 2         | A₁ = 1<br>A₂ = 1            | x₁ = −1/√3<br>x₂ = 1/√3        | 3   |
| 3         | A₁ = 5/9<br>A₂ = 8/9<br>A₃ = 5/9 | x₁ = −√(3/5)<br>x₂ = 0<br>x₃ = √(3/5) | 5   |

### Задача 3
Да се пресметне приближено стойността на интеграла
$\displaystyle I=\int_{0}^{0.8} f(x) ~dx$, където $f(x) = 0.2 + 25x - 200x^2 + 675x^3 - 900x^4 + 400x^5$, като се използват формулите на Гаус–Льожандър с 2 и 3 възела с подходяща смяна на променливите. Да се сравни резултатът с точната стойност. Да се сравни със стойностите, получени по формулите
на трапеците и на Симпсън.

```python
import sympy as sp
import numpy as np

a = 0
b = 0.8
def f(x):
    return 0.2 + 25*x - 200*x**2 + 675*x**3 - 900*x**4 + 400*x**5

x = sp.symbols('x')
print(sp.integrate(f(x), (x,0,0.8)))

t = sp.symbols('t')
def phi(t):
    return ((b-a)/2)* f((b-a)/2 + ((b-a)/2) * t)

print(sp.integrate(phi(t), (t,-1,1)))
print("Gauss-Legendre 2 nodes: ", phi(-np.sqrt(3) / 3) + phi (np.sqrt(3) / 3))
print("Gauss-Legendre 3 nodes: ", 5 / 9 * phi(-np.sqrt(3 / 5)) + 8 / 9 * phi(0) + 5 / 9 * phi(np.sqrt(3 / 5)))
```
