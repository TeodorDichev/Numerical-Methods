**Задача 8.**  
Да се пресметне приблизителната стойност на интеграла
$$
\int_{-1}^{1} (-x^4 + 1)\,dx
$$
като се използва **съставната квадратурна формула на трапеците** с **5 равноотдалечени възела** в интервала \([-1,1]\).

```python
import numpy as np
import sympy as sp

x = sp.Symbol('x')

def f(x):
    return -x**4 + 1

# d_f = sp.diff(f(x), x)
# d_d_f = sp.diff(d_f, x)

# 0 максимизира грешката в интервала -1, 1
# казано е приблизителната стойност това е ненужно
# epsilon = 0
# def err():
#     return -(d_d_f(epsilon)/(12*n**2))*(b-a)**3

a = -1
b = 1
n = 5
# 5 възела => 4 интервала
# ако бяха казани 5 интервала щяхме да вземем 5+1 възела
x_mesh = np.linspace(a, b, n)
app_integral_value = 0

for i in range(1, x_mesh.size):
    app_integral_value += f(x_mesh[i-1]) + f(x_mesh[i])
    
# n - 1 защото са 4 интервала
app_integral_value *= (b-a)/(2*(n-1))

print(app_integral_value)
print(sp.integrate(f(x), (x, -1, 1)).evalf())
```

**Задача 9.**  
Да се намери приблизителната стойност на интеграла
$$
\int_{0}^{0.4} f(x)\,dx
$$
използвайки **съставната квадратурна формула на Симпсън**, като са известни само следните стойности на функцията:

| x     | 0   | 0.1 | 0.2 | 0.3 | 0.4 |
|-------|-----|-----|-----|-----|-----|
| f(x)  | 1   | 8   | 4   | 3.5 | 5   |

```python
import numpy as np
import sympy as sp

x_mesh = np.array([0,0.1,0.2,0.3,0.4])
y = np.array([1,8,4,3.5,5])

a = 0
b = 0.4
n = x_mesh.size

app_integral_value = 0

for i in range(1, x_mesh.size):
    app_integral_value += y[i-1] + 4*f((x_mesh[i-1] + x_mesh[i])/2) + y[i]
    
# n - 1 защото са 4 интервала
app_integral_value *= (b-a)/(6*(n-1))

print(app_integral_value)
```

**Задача 10.**  
Нека f(x) = 1 - cotg(x).  
Да се намери **n**, така че n-тата съставна квадратурна формула (правоъгълници / трапеци / Симпсън) да приближава  

$$
\int_{-2}^{-1} f(x)\,dx
$$

с грешка, **не по-голяма от** epsilon = 10^-5.  
Да се пресметне **приблизителното интегрално приближение**.


```python
# правоъгълници
import sympy as sp
import numpy as np

x = sp.Symbol('x')
b = -1
a = -2

def f(x):
    return 1 - sp.cos(x)/sp.sin(x)

def f_num(x):
    return 1 - np.cos(x)/np.sin(x)

# намираме втората производна
dd_f = sp.diff(sp.diff(f(x), x), x)
print(dd_f)

# трябва да сметнем n
# може на хартия може и тук
space = np.linspace(-2, -1, 1000)
dd_f_num = sp.lambdify(x, dd_f, 'numpy')
max_f2 = np.max(np.abs(dd_f_num(space)))

epsilon = 1e-5

n = int(np.ceil(np.sqrt((b-a)**3 * max_f2 / (24 * epsilon))))
print(int(n))
print(f"{(b-a)**3 * max_f2 / (24 * n ** 2):.10f}")

# стандартно смятаме интегралното приближение
x_mesh = np.linspace(a, b, n + 1)
app_val = 0

for i in range(1, n + 1):
    app_val += f_num((x_mesh[i-1] + x_mesh[i])/2)
    
app_val *= (b-a)/n
print(app_val) 
print(sp.integrate(f(x), (x, a, b)).evalf()) 
```

```python
# трапеци
import sympy as sp
import numpy as np

x = sp.Symbol('x')
b = -1
a = -2

def f(x):
    return 1 - sp.cos(x)/sp.sin(x)

def f_num(x):
    return 1 - np.cos(x)/np.sin(x)

# намираме втората производна
dd_f = sp.diff(sp.diff(f(x), x), x)
print(dd_f)

# трябва да сметнем n
# може на хартия може и тук
space = np.linspace(-2, -1, 1000)
dd_f_num = sp.lambdify(x, dd_f, 'numpy')
max_f2 = np.max(np.abs(dd_f_num(space)))

epsilon = 1e-5

# без минус, взима се модулът
n = int(np.ceil(np.sqrt((b-a)**3 * max_f2 / (12 * epsilon))))
print(int(n))
print(f"{(b-a)**3 * max_f2 / (12 * n ** 2):.10f}")

# стандартно смятаме интегралното приближение
x_mesh = np.linspace(a, b, n + 1)
app_val = 0

for i in range(1, n + 1):
    app_val += f_num(x_mesh[i-1]) + f_num(x_mesh[i])
    
app_val *= (b-a)/(2*n)
print(app_val) 
print(sp.integrate(f(x), (x, a, b)).evalf()) 
```

```python
# симпсън
import sympy as sp
import numpy as np

x = sp.Symbol('x')
b = -1
a = -2

def f(x):
    return 1 - sp.cos(x)/sp.sin(x)

def f_num(x):
    return 1 - np.cos(x)/np.sin(x)

# намираме четвъртата производна
dddd_f = sp.diff(sp.diff(sp.diff(sp.diff(f(x), x), x), x), x)
print(dddd_f)

# трябва да сметнем n
# може на хартия може и тук
space = np.linspace(-2, -1, 1000)
dddd_f_num = sp.lambdify(x, dddd_f, 'numpy')
max_f4 = np.max(np.abs(dddd_f_num(space)))

epsilon = 1e-5

# без минус, взима се модулът
n = int(np.ceil(np.power((b-a)**5 * max_f4 / (2880 * epsilon), 1/4)))
print(int(n))
print(f"error: {(b-a)**5 * max_f4 / (2880 * n ** 4):.10f}")

# стандартно смятаме интегралното приближение
x_mesh = np.linspace(a, b, n + 1)
app_val = 0

for i in range(1, n + 1):
    app_val += f_num(x_mesh[i-1]) + 4 * f_num((x_mesh[i-1] + x_mesh[i])/2)+ f_num(x_mesh[i])
    
app_val *= (b-a)/(6*n)
print("approximate: ", app_val) 
print("real: ",sp.integrate(f(x), (x, a, b)).evalf()) 
```

Гаус за 3 възела:
И приближено с нея да се сметне:
$$
\int_{0}^{1} \big(e^{\sin(x)} - 2x^2\big)\,dx
$$

```python
import sympy as sp
import numpy as np

x = sp.symbols('x')
x1 = sp.symbols('x1')
x2 = sp.symbols('x2')
x3 = sp.symbols('x3')
A1 = sp.symbols('A1')
A2 = sp.symbols('A2')
A3 = sp.symbols('A3')

sol = sp.solve(
    [
        sp.Eq(sp.integrate(1, (x,-1,1)), A1 + A2 + A3),
        sp.Eq(sp.integrate(x, (x,-1,1)), A1*x1 + A2*x2 + A3*x3),
        sp.Eq(sp.integrate(x**2, (x,-1,1)), A1*x1**2 + A2*x2**2 + A3*x3**2),
        sp.Eq(sp.integrate(x**3, (x,-1,1)), A1*x1**3 + A2*x2**3 + A3*x3**3),
        sp.Eq(sp.integrate(x**4, (x,-1,1)), A1*x1**4 + A2*x2**4 + A3*x3**4),
        sp.Eq(sp.integrate(x**5, (x,-1,1)), A1*x1**5 + A2*x2**5 + A3*x3**5)
    ]
)

def f(x):
    return sp.exp(sp.sin(x)) - 2*x**2

def nf(x):
    return np.exp(np.sin(x)) - 2*x**2

def appr():
    return sol[0][A1]*nf(float(sol[0][x1])) + sol[0][A2]*nf(float(sol[0][x2])) + sol[0][A3]*nf(float(sol[0][x3]))
    
print(appr())
print(sp.integrate(f(x), (x,0,1)).evalf())

```
