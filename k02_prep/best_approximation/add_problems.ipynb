–î–∞ —Å–µ –Ω–∞–º–µ—Ä–∏ –ø–æ–ª–∏–Ω–æ–º –æ—Ç –≤—Ç–æ—Ä–∞ —Å—Ç–µ–ø–µ–Ω –ø–æ –º–µ—Ç–æ–¥–∞ –Ω–∞ –Ω–∞–π-–¥–æ–±—Ä–æ —Å—Ä–µ–¥–Ω–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ 
[0,ùúã/2] –ø—Ä–∏ —Ç–µ–≥–ª–æ–≤–∞—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è ùúá(ùë•)=1 –∑–∞ —Ñ—É–Ω–∫—Ü–∏—è—Ç–∞ f(x)=sinx.

–î–∞ —Å–µ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–∞ –≥—Ä–∞—Ñ–∏—á–Ω–æ –∏ –¥–∞ —Å–µ –∏–∑—á–∏—Å–ª–∏ —Å—Ä–µ–¥–Ω–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ—Ç–æ —Ä–∞–∑—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –ø–æ–ª–∏–Ω–æ–º–∞ –∏ —Ñ—É–Ω–∫—Ü–∏—è—Ç–∞.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# g(x) = a*x**2 + b*x + c

a,b,c,x = sp.symbols('a,b,c,x')

def f(x):
    return sp.sin(x)

def g(a,b,c,x):
    return a*x**2 + b*x + c

def integral(a,b,c):
    return sp.integrate((f(x) - g(a,b,c,x))**2, (x,0,np.pi/2))

equations = [sp.Eq(sp.diff(integral(a,b,c), a), 0),
             sp.Eq(sp.diff(integral(a,b,c), b), 0),
             sp.Eq(sp.diff(integral(a,b,c), c), 0)]

sol = sp.solve(equations) # print it if needed

f_num = sp.lambdify(x, f(x), 'numpy')
g_num = sp.lambdify(x, g(sol[a], sol[b], sol[c], x), 'numpy')

# —Å—Ä–µ–¥–Ω–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ —Ä–∞–∑—Å—Ç–æ—è–Ω–∏–µ
print(sp.sqrt(sp.integrate((f(x) - g(sol[a], sol[b], sol[c], x))**2, (x,0,np.pi/2))))

space = np.linspace(0, np.pi/2)
plt.plot(space, f_num(space), space, g_num(space))
plt.show()
```

–î–∞ —Å–µ –Ω–∞–º–µ—Ä–∏ –ø–æ–ª–∏–Ω–æ–º –æ—Ç —Ç—Ä–µ—Ç–∞ —Å—Ç–µ–ø–µ–Ω –∑–∞ –Ω–∞–π-–¥–æ–±—Ä–æ —Å—Ä–µ–¥–Ω–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ [0,1] –ø—Ä–∏ —Ç–µ–≥–ª–æ–≤–∞ —Ñ—É–Ω–∫—Ü–∏—è Œº(x)=1 –∑–∞ —Ñ—É–Ω–∫—Ü–∏—è—Ç–∞ f(x)=e**x + 1.

–î–∞ —Å–µ –∏–ª—é—Å—Ç—Ä–∏—Ä–∞ –≥—Ä–∞—Ñ–∏—á–Ω–æ –ø–æ–ª—É—á–µ–Ω–æ—Ç–æ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –∏ –¥–∞ —Å–µ –Ω–∞–º–µ—Ä–∏ —Å—Ä–µ–¥–Ω–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ—Ç–æ —Ä–∞–∑—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –ø–æ–ª–∏–Ω–æ–º–∞ –∏ —Ñ—É–Ω–∫—Ü–∏—è—Ç–∞.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

a,b,c,d,x = sp.symbols('a,b,c,d,x')

def f(x):
    return sp.exp(x) + 1

def g(a,b,c,d,x):
    return a*x**3 + b*x**2 + c*x + d

def integral(a,b,c,d):
    return sp.integrate((f(x) - g(a,b,c,d,x))**2, (x,0,1))

equations = [sp.Eq(sp.diff(integral(a,b,c,d), a), 0),
             sp.Eq(sp.diff(integral(a,b,c,d), b), 0),
             sp.Eq(sp.diff(integral(a,b,c,d), c), 0),
             sp.Eq(sp.diff(integral(a,b,c,d), d), 0)]

sol = sp.solve(equations)

# —Å—Ä–µ–¥–Ω–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ —Ä–∞–∑—Å—Ç–æ—è–Ω–∏–µ
print(sp.sqrt(sp.integrate((f(x) - g(sol[a],sol[b],sol[c],sol[d],x))**2, (x,0,1))).evalf())

f_num = sp.lambdify(x, f(x), "numpy")
g_num = sp.lambdify(x, g(sol[a],sol[b],sol[c],sol[d],x), "numpy")

space = np.linspace(0,2,1000)
plt.plot(space, f_num(space), space, g_num(space))
plt.show()
```
